# Author Overlap Matrix

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def author_overlap(df1, df2):
    authors_1 = set(df1['author'].dropna().unique())
    authors_2 = set(df2['author'].dropna().unique())
    overlap = authors_1.intersection(authors_2)
    percentage = len(overlap) / len(authors_1) * 100 if len(authors_1) > 0 else 0
    return len(overlap), percentage

# Load datasets
reddit_paris_olympics = pd.read_csv("datasets/reddit/paris_olympics.csv")
reddit_tokyo_olympics = pd.read_csv("datasets/reddit/tokyo_olympics.csv")
reddit_qatar_world_cup = pd.read_csv("datasets/reddit/qatar_world_cup.csv")
reddit_russia_world_cup = pd.read_csv("datasets/reddit/russia_world_cup.csv")

olymlics = pd.concat([reddit_paris_olympics, reddit_tokyo_olympics])
world_cup = pd.concat([reddit_qatar_world_cup, reddit_russia_world_cup])

dfs = [olymlics, world_cup]
labels = ["Olympics", "World Cup"]

dfs = [reddit_tokyo_olympics, reddit_paris_olympics, reddit_russia_world_cup, reddit_qatar_world_cup, olymlics, world_cup]
labels = ["Paris\nOlympics", "Tokyo\nOlympics", "Qatar\nWorld Cup", "Russia\nWorld Cup", "Olympics\nCombined", "World Cup\nCombined"]

overlap_matrix = np.zeros((len(dfs), len(dfs)))
annotations = np.empty_like(overlap_matrix, dtype=object)

for i, df1 in enumerate(dfs):
    for j, df2 in enumerate(dfs):
        raw, pct = author_overlap(df1, df2)
        overlap_matrix[i, j] = pct
        annotations[i, j] = f"{raw}\n({pct:.2f}%)"

plt.figure(figsize=(8, 6))
sns.heatmap(overlap_matrix, xticklabels=labels, yticklabels=labels,
            annot=annotations, fmt="", cmap="coolwarm")
plt.title("Author Overlap Matrix (Reddit)", fontsize=20)
plt.ylabel("Event")
plt.xlabel("Event")
plt.yticks(rotation=0)
plt.show()
```
