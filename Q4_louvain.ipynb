# Community Detection

```python
import matplotlib.pyplot as plt
import networkx as nx
from collections import Counter

def plot_community_distribution(graphml_path, title):

    G = nx.read_graphml(graphml_path)
    communities = [int(data['community']) for _, data in G.nodes(data=True) if 'community' in data]

    community_counts = Counter(communities)
    
    plt.figure(figsize=(15, 3))
    plt.bar(community_counts.keys(), community_counts.values(), color='skyblue')
    plt.xlabel('Community ID')
    plt.ylabel('Number of Users')
    plt.title(f'Distribution of community size - {title} (Total {len(community_counts)} communities)', fontsize=16)
    plt.xticks(list(community_counts.keys()))
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.show()
```

```python
plot_community_distribution("graphs/reddit/tokyo_olympics_louvain.graphml", "Tokyo Olympics")
```

```python
plot_community_distribution("graphs/reddit/paris_olympics_louvain.graphml", "Paris Olympics")
```
